```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
import tensorflow as tf
from tensorflow.keras.datasets import mnist
```

```python
import warnings 
warnings.filterwarnings("ignore")
```

```python
(train_images , train_labels),(test_images,test_labels)=mnist.load_data()
```

```python
# Normalize the pixel values to the range [0, 1]
train_images = train_images.astype('float32')/255.0
test_images = test_images.astype('float32') / 255.0
```

```python
# Convert labels to one-hot encoding
train_labels = tf.keras.utils.to_categorical(train_labels,10)
test_labels = tf.keras.utils.to_categorical(test_labels,10)
```

```python
def display_sample(images,labels,num_samples=5):
    plt.figure(figsize=(10,5))
    for i in range(num_samples):
        plt.subplot(1,num_samples,i+1)
        plt.imshow(images[i],cmap="gray")
        plt.title(f'Lable:{labels[i]}')
        plt.axis('off')
    plt.show()
```

```python
display_sample(train_images,test_images)
```

```python
print("Training image shape:",train_images.shape)
print("Training labels shape:",train_labels.shape)
print("Testing image shape:",test_images.shape)
print("Testing labels shape:",test_labels.shape)
```

# Building CNN

```python
# Initialising the CNN
cnn = tf.keras.models.Sequential()
```

```python
# Step 1 = Convolution layer
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size=3,activation="relu" , input_shape=[28, 28, 1]))
```

```python
# Step 2 - MaxPooling Layer
cnn.add(tf.keras.layers.MaxPooling2D(pool_size=2,strides=2))
```

```python
# Step 3 = Convolution Layer
cnn.add(tf.keras.layers.Conv2D(filters=64,kernel_size=3,activation="relu"))
```

```python
# Step 4 - MaxPooling Layer
cnn.add(tf.keras.layers.MaxPooling2D(pool_size=2,strides=2))
```

```python
# Step 5 - Flattening
cnn.add(tf.keras.layers.Flatten())
```

```python
# Step 6 - Fully Connected 
cnn.add(tf.keras.layers.Dense(units=64,activation="relu"))
```

```python
# Step 7 - Output Layer
cnn.add(tf.keras.layers.Dense(units=10,activation="softmax"))
```

# Model Compilation

```python
cnn.compile(optimizer="adam",
           loss='categorical_crossentropy',
           metrics=['accuracy'])
```
